```python
# MCD record
roc_train_mcd = []
prn_train_mcd = []
roc_test_mcd = []
prn_test_mcd = []
roc_diff_mcd = []
prn_diff_mcd = []
```

```python
# CBLOF record
roc_train_cblof = []
prn_train_cblof = []
roc_test_cblof = []
prn_test_cblof = []
roc_diff_cblof = []
prn_diff_cblof = []
```

```python
winediff = open("skinBenchmarkResults.txt","r")
lines = winediff.readlines()
```

```python
def get_roc_prn(line):
    item = line.split(" - ")
    roc = float(item[0].split(":")[1])
    prn = float(item[1].split(":")[1])
    return roc, prn
```

```python
# clf flag
# 0 ~ 7
clf_num = 0
clf_name = ['KNN', 'avg KNN', 'PCA', 'MCD', 'CBLOF','FB', 'HBOS', 'ISOF']
# list flag
# 0 ~ 2
train_test_diff = 0
```

```python
print("Reading file...")
for line in lines:
    if 'roc' in line:
        roc, prn = get_roc_prn(line)
        # train set results
        if train_test_diff == 0:
            # add data to list
            if clf_num == 0:
                roc_train_knn.append(roc)
                prn_train_knn.append(prn)
            elif clf_num == 1:
                roc_train_avgknn.append(roc)
                prn_train_avgknn.append(prn)
            elif clf_num == 2:
                roc_train_pca.append(roc)
                prn_train_pca.append(prn)
            elif clf_num == 3:
                roc_train_mcd.append(roc)
                prn_train_mcd.append(prn)
            elif clf_num == 4:
                roc_train_cblof.append(roc)
                prn_train_cblof.append(prn)
            elif clf_num == 5:
                roc_train_fb.append(roc)
                prn_train_fb.append(prn)
            elif clf_num == 6:
                roc_train_hbos.append(roc)
                prn_train_hbos.append(prn)
            elif clf_num == 7:
                roc_train_isof.append(roc)
                prn_train_isof.append(prn)
            # change mode
            train_test_diff += 1

        # test set results
        elif train_test_diff == 1:
            if clf_num == 0:
                roc_test_knn.append(roc)
                prn_test_knn.append(prn)
            elif clf_num == 1:
                roc_test_avgknn.append(roc)
                prn_test_avgknn.append(prn)
            elif clf_num == 2:
                roc_test_pca.append(roc)
                prn_test_pca.append(prn)
            elif clf_num == 3:
                roc_test_mcd.append(roc)
                prn_test_mcd.append(prn)
            elif clf_num == 4:
                roc_test_cblof.append(roc)
                prn_test_cblof.append(prn)
            elif clf_num == 5:
                roc_test_fb.append(roc)
                prn_test_fb.append(prn)
            elif clf_num == 6:
                roc_test_hbos.append(roc)
                prn_test_hbos.append(prn)
            elif clf_num == 7:
                roc_test_isof.append(roc)
                prn_test_isof.append(prn)
            # change mode
            train_test_diff += 1
        
        # diff set results
        elif train_test_diff == 2:
            if clf_num == 0:
                roc_diff_knn.append(roc)
                prn_diff_knn.append(prn)
                clf_num += 1
            elif clf_num == 1:
                roc_diff_avgknn.append(roc)
                prn_diff_avgknn.append(prn)
                clf_num += 1
            elif clf_num == 2:
                roc_diff_pca.append(roc)
                prn_diff_pca.append(prn)
                clf_num += 1
            elif clf_num == 3:
                roc_diff_mcd.append(roc)
                prn_diff_mcd.append(prn)
                clf_num += 1
            elif clf_num == 4:
                roc_diff_cblof.append(roc)
                prn_diff_cblof.append(prn)
                clf_num += 1
            elif clf_num == 5:
                roc_diff_fb.append(roc)
                prn_diff_fb.append(prn)
                clf_num += 1
            elif clf_num == 6:
                roc_diff_hbos.append(roc)
                prn_diff_hbos.append(prn)
                clf_num += 1
            elif clf_num == 7:
                roc_diff_isof.append(roc)
                prn_diff_isof.append(prn)
                clf_num = 0
            # change mode
            train_test_diff = 0
print("Read done.")
```

```python
def show_roc3(train, test, diff):
    plt.figure(figsize=(16,12))
    x = range(len(train))

    bar1 = plt.bar(x = [i - 0.2 for i in x], height = train, width = 0.2, alpha = 0.8, color = 'lightblue',label = 'Train')
    bar2 = plt.bar(x = [i for i in x], height = test, width = 0.2, alpha = 0.8,color = 'lightgreen',label = 'Test')
    bar3 = plt.bar(x = [i + 0.2 for i in x], height = diff, width = 0.2, alpha = 0.8,color = 'pink',label = 'Diff')

    plt.ylim(0,1)
    plt.title('ROC指标')
    plt.xlabel('Benchmark')
    plt.ylabel('ROC value')
    plt.legend()
    plt.show()
```

```python
def show_prn3(train, test, diff):
    plt.figure(figsize=(16,12))
    x = range(len(train))

    bar1 = plt.bar(x = [i - 0.2 for i in x], height = train, width = 0.2, alpha = 0.8, color = 'lightblue',label = 'Train')
    bar2 = plt.bar(x = [i for i in x], height = test, width = 0.2, alpha = 0.8,color = 'lightgreen',label = 'Test')
    bar3 = plt.bar(x = [i + 0.2 for i in x], height = diff, width = 0.2, alpha = 0.8,color = 'pink',label = 'Diff')

    plt.ylim(0,1)
    plt.title('PRN指标')
    plt.xlabel('Benchmark')
    plt.ylabel('PRN value')
    plt.legend()
    plt.show()
```

```python
def show_all3(roc_train, roc_test, roc_diff, prn_train, prn_test, prn_diff):
    plt.figure(figsize=(13,10))
    x = range(len(roc_train))
    
    plt.subplot(321)
    plt.plot(x,roc_train,c='lightblue')
    plt.ylim(0,1)
    plt.title('ROC TRAIN')
    plt.xlabel('Benchmark')
    plt.ylabel('ROC value')
    
    plt.subplot(322)
    plt.plot(x,prn_train,c='lightpink')
    plt.ylim(0,1)
    plt.title('PRN TRAIN')
    plt.xlabel('Benchmark')
    plt.ylabel('PRN value')
    
    plt.subplot(323)
    plt.plot(x,roc_test,c='lightblue')
    plt.ylim(0,1)
    plt.title('ROC TEST')
    plt.xlabel('Benchmark')
    plt.ylabel('ROC value')
    
    plt.subplot(324)
    plt.plot(x,prn_test,c='lightpink')
    plt.ylim(0,1)
    plt.title('PRN TEST')
    plt.xlabel('Benchmark')
    plt.ylabel('PRN value')
    
    plt.subplot(325)
    plt.plot(x,roc_diff,c='lightblue')
    plt.ylim(0,1)
    plt.title('ROC DIFF')
    plt.xlabel('Benchmark')
    plt.ylabel('ROC value')
    
    plt.subplot(326)
    plt.plot(x,prn_diff,c='lightpink')
    plt.ylim(0,1)
    plt.title('PRN DIFF')
    plt.xlabel('Benchmark')
    plt.ylabel('PRN value')
    
    plt.tight_layout()
    
    plt.show()
```

### 我们对每一种离群点分类器效果进行单独可视化

```python
show_all3(roc_train_knn, roc_test_knn, roc_diff_knn, prn_train_knn, prn_test_knn, prn_diff_knn)
```

KNN的ROC参数在每个benchmark上都比较高，大部分都保持在0.9之上，ROC DIFF的效果比单个benchmark稍微差一点，应是全数据集的离群点与benchmark中的标定不同；
